2D Array Slicing




```python
arr = np.array([[10,20,30], 
                [40,50,60]])
print(arr[0,0])
print(arr[1,2])
print(arr[:, 2])
print(arr[1:,])
```

Mathematical Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("addition:", a+b)
print("substraction:", a-b)
print("multiplication:", a*b)
print("division:", a/b)
print("mean of a:", np.mean(a))
print("mean of b: ", np.mean(b))
print("max of a : ", np.max(a))
print("max of b : ", np.max(b))
```

Reshape and Flatten

```python
arr = np.arange(1,10)

matrix = arr.reshape(3,3)
print("Matrix", matrix)

flat = matrix.flatten()
print("Flattened", flat)

arr2 = np.array([[1,2,3,4],
                 [5,6,7,8]])
print("Matrix 2: ", arr2)
print("Matrix to normal single line", arr2.flatten())
```
