```python
# Create Data  Frame
import pandas as pd
data ={
    'Name': ['Alice', 'Jon', 'Bob'],
    'Age': [25,30,35],
    'City': ['Delhi', 'Mumbai', 'Bangalore']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Explore Data
print(df.head())
print( df.info(), "Information about data frame")

print(df.describe())
```

Selecting Column and Row

```python
print(df['Name'])
print("\n")
print(df[['Name', 'City']])
print("\n")
print(df.iloc[0])
```

Filtering Data

```python
print(df[df['Age']> 25])
```

Add or Modify Columns

```python
df['Country'] = 'India'

df['After_5_years'] = df['Age'] + 5

print(df)
```

Sorting and Grouping

```python
print(df.sort_values(by = 'Age', ascending=False))


print("\n")
print(df.groupby('City')['Age'].mean())
```

Handle Missing Values

```python
df.loc[1, 'City'] = None     # make one value missing
print(df.isnull())
print('\n')
df.fillna('Unknown', inplace=True)
print(df)
```

Merge / Join

```python
df1 = pd.DataFrame({
    'ID': [1,2,3],
    'Name': ['A','B','C']
    
})

df2 = pd.DataFrame({
    'ID': [1,2,3],
    'Score': [50,60,70]
    
})

print(pd.merge(df1,df2, on='ID'))
```

```python
# simpple Visualization

import matplotlib.pyplot as plt

df['Age'].plot(kind='bar', title='Age of people')

plt.show()

df2['Score'].plot(kind='bar', title='Score')
```
